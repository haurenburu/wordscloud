```python
#!pip install newspaper3k
#!pip install wordcloud
from newspaper import Article
import pandas as pd

#article = Article('https://en.wikipedia.org/wiki/Magic:_The_Gathering')
article = Article('https://magic.wizards.com/en/mtgarena')
article.download()
article.parse()

from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from PIL import Image


def create_word(string):
    mask = np.array(Image.open("cloud.jpg"))
    cloud = WordCloud(background_color = "#fff", max_font_size=50, max_words = 200, mask = mask, stopwords = set(STOPWORDS))
    cloud.generate(string)
    cloud.to_file("result.png")
    plt.figure()
    plt.imshow(cloud, interpolation='bilinear')
    plt.axis('off')

create_word(article.text)
```
